```python
import pandas as pd
import glob
import os

print("=== PREPROCESAMIENTO COLDTRACK - UNIÓN DE 18 DÍAS ===\n")

# ====================== RUTA EXACTA (AJUSTADA A TU CASO) ======================
# Opción 1: Ruta relativa (la más común)
ruta_datos = "datos_coldtrack_diarios"

# Opción 2: Si no funciona, descomenta la siguiente línea (ruta absoluta):
# ruta_datos = "/Users/josea/Documents/pruebas-json/datos_coldtrack_diarios"

# =============================================================================

# Verificar si la carpeta existe
if not os.path.exists(ruta_datos):
    print(f"❌ ERROR: La carpeta '{ruta_datos}' no existe.")
    print("Lista de carpetas/archivos en la ubicación actual:")
    print(os.listdir('.'))
else:
    # Buscar todos los archivos CSV
    archivos = sorted(glob.glob(os.path.join(ruta_datos, "datos_refrigerador_*.csv")))
    
    print(f"Se encontraron {len(archivos)} archivos CSV en la carpeta '{ruta_datos}'\n")
    
    if len(archivos) == 0:
        print("No se encontraron archivos con el patrón 'datos_refrigerador_*.csv'")
        print("Archivos en la carpeta:")
        print(os.listdir(ruta_datos))
    else:
        # Mostrar información de los archivos
        print("Primeros 3 y últimos 2 archivos:")
        for f in archivos[:3]:
            print(f"   → {os.path.basename(f)}")
        for f in archivos[-2:]:
            print(f"   → {os.path.basename(f)}")
        
        print("\nUniendo los 18 archivos... (esto puede tardar unos segundos)")
        
        # Unir todos los CSV
        df_list = []
        for f in archivos:
            temp_df = pd.read_csv(f)
            df_list.append(temp_df)
        
        df = pd.concat(df_list, ignore_index=True)
        
        print("\n" + "="*65)
        print("=== DATASET COMBINADO EXITOSAMENTE ===")
        print("="*65)
        print(f"Total de registros: {df.shape[0]:,}")
        print(f"Total de columnas:  {df.shape[1]}")
        print(f"Primer registro : {df['Timestamp'].iloc[0]}")
        print(f"Último registro : {df['Timestamp'].iloc[-1]}")
        
        print("\nDistribución de flag_mantenimiento:")
        print(df['flag_mantenimiento'].value_counts())
        print("\nPorcentajes:")
        print((df['flag_mantenimiento'].value_counts(normalize=True) * 100).round(2))
        
        # Guardar el dataset completo
        df.to_csv("coldtrack_dataset_completo.csv", index=False)
        print("\n✅ Archivo guardado correctamente: coldtrack_dataset_completo.csv")
        print(f"Tamaño aproximado: {os.path.getsize('coldtrack_dataset_completo.csv') / (1024*1024):.2f} MB")
```
